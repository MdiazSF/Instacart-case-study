#6 Q) The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

```python
#split data into groups based on user_id, and transform prices to generate average_price for each user
ords_prods_merge['average_price']= ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check for accuracy
ords_prods_merge.head(60)
```

```python
#derive column conditions
ords_prods_merge.loc[ords_prods_merge['average_price']<10, 'spending_flag']= 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price']>=10, 'spending_flag']= 'High spender'
```

```python
#frequency of spending_flag
ords_prods_merge['spending_flag'].value_counts(dropna= False)
```

#7 Q) In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. 

Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column. The criteria for the flag should be as follows:
If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer.
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer.
If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer.


```python
#split data into groups based on user_ids, and transform days_since_prior_order to generate median_order_frequency
ords_prods_merge['median_order_frequency']= ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check for accuracy
ords_prods_merge.head(60)
```

```python
#derive column conditions
ords_prods_merge.loc[ords_prods_merge['median_order_frequency']>20, 'order_frequency_flag']= 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency']>10) & (ords_prods_merge['median_order_frequency']<=20), 'order_frequency_flag']= 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_frequency']<=10, 'order_frequency_flag']= 'Frequent customer'
```

```python
#frequency of order_frequency_flag
ords_prods_merge['order_frequency_flag'].value_counts(dropna= False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'spending_flag', 'order_frequency_flag']].head(60)
```
